```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score

# models
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# explainability
import shap
from sklearn.metrics import plot_confusion_matrix
```

```python
file = 'stats_REDUCEDTOSUPPORT_2022-03-14_500records.csv'
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df = pd.read_csv('../data/' + file)
df.head()
```

```python
df.shape
```

# Dataframe Manipulation

```python
columns = df.select_dtypes(include='number').drop(['internalID'], axis=1).columns
min_max = MinMaxScaler()
df[columns] = min_max.fit_transform(df[columns])

df['damageDealtToObjectives'] = df['damageDealtToObjectives'] - df['damageDealtToTurrets']

df = df.drop([
    'matchID', 
    'date', 
    'gameType', 
    'championTransform', 
    'summonerName', 
    'item0', 
    'item1', 
    'item2',
    'item3', 
    'item4', 
    'item5', 
    'item6', 
    'summoner1Id', 
    'summoner2Id',
    'internalID',
    
#     'turretTakedowns',
#     'damageDealtToObjectives',
#     'bountyLevel',
#     'totalTimeSpentDead',
    
#     'damageToTurretsRanking',
    
#     'turretsLost',
#     'inhibitorsLost',
#     'turretKills',
#     'nexusLost',
#     'damageDealtToTurrets',
#     'nexusTakedowns',
#     'inhibitorTakedowns',
    'gameDuration',
    'queueID'
], axis=1).replace(False, 0).replace(True, 1)
df.dtypes[df.dtypes == 'object'].index

df = df.dropna(axis=1, how='any')
```

# ML Preprocessing

```python
# X = pd.get_dummies(df, columns=['championName', 'individualPosition']).drop(columns=['win'])
X = pd.get_dummies(df, columns=['championName']).drop(columns=['win'])
y = df['win']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, train_size=0.75)
```

# Model Training

```python
forest = RandomForestClassifier(n_estimators=1000, random_state=42, n_jobs=-1)
forest.fit(X_train, y_train)
print('Score:', forest.score(X_test, y_test))

forest_importance = forest.feature_importances_

feature_importance_list = []
for feature_index, feature_score in enumerate(forest_importance):
    feature_importance_list.append((X.columns.tolist()[feature_index], round(feature_score * 100, 2)))
    
forest_feature_df = pd.DataFrame(feature_importance_list, columns=['feature', 'percent_importance']).sort_values('percent_importance', ascending=False)
    
# plot feature importance
plt.bar([x for x in range(len(forest_importance))], forest_importance)
plt.title('First Forest Importances')
plt.xlabel('Feature #')
plt.ylabel('% Importance')
plt.show()

forest_feature_df.head(30)
```

```python
score = 0
cutoff = 0
for i in np.arange(0, 3.5, 0.5):
    X_train_trimmed = X_train.drop(forest_feature_df[forest_feature_df['percent_importance'] < i]['feature'].tolist(), axis=1)
    X_test_trimmed = X_test.drop(forest_feature_df[forest_feature_df['percent_importance'] < i]['feature'].tolist(), axis=1)

    deforestation = RandomForestClassifier(n_estimators=1000, random_state=42, n_jobs=-1)
    deforestation.fit(X_train_trimmed, y_train)
    print('Score:', deforestation.score(X_test_trimmed, y_test), i)
    if deforestation.score(X_test_trimmed, y_test) >= score:
        score = deforestation.score(X_test_trimmed, y_test)
        cutoff = i

X_train_trimmed = X_train.drop(forest_feature_df[forest_feature_df['percent_importance'] < cutoff]['feature'].tolist(), axis=1)
X_test_trimmed = X_test.drop(forest_feature_df[forest_feature_df['percent_importance'] < cutoff]['feature'].tolist(), axis=1)

deforestation = RandomForestClassifier(n_estimators=1000, random_state=42, n_jobs=-1)
deforestation.fit(X_train_trimmed, y_train)
print(f"Best Cutoff: {cutoff}, Score: {deforestation.score(X_test_trimmed, y_test)}")
        
deforestation_importance = deforestation.feature_importances_

feature_importance_list = []
for feature_index, feature_score in enumerate(deforestation_importance):
    feature_importance_list.append((X.columns.tolist()[feature_index], round(feature_score * 100, 2)))
    
deforestation_feature_df = pd.DataFrame(feature_importance_list, columns=['feature', 'percent_importance']).sort_values('percent_importance', ascending=False)
    
# plot feature importance
deforestation_feature_df.sort_values('percent_importance').plot(x='feature', y='percent_importance', kind='barh', figsize=(10, 5))
plt.title('Deforestation Importances')
plt.ylabel('Feature')
plt.xlabel('% Importance')
plt.show()

deforestation_feature_df.head(10)
```

```python
drill = pd.read_csv('../data/' + file)
drill = drill[['assists', 'win', 'matchID']]
drill['win'] = drill['win'].map({False: 0, True: 1})
# plt.figure(dpi=1200)
drill.plot.scatter(x='matchID', y='assists', figsize=(20, 5), s=100, c='win', cmap='bwr_r')
plt.xticks(rotation=90)
plt.show()
```

```python
plot_confusion_matrix(
    deforestation,
    X_test_trimmed,
    y_test,
    display_labels=y.map({0: 'Loss', 1: 'Win'}).unique(),
    cmap='cividis'
)
plt.title('Deforestation Confusion Matrix')
plt.grid(False)
plt.xticks(rotation=0)
plt.show()
```

## XGBoost

```python
model = XGBClassifier(use_label_encoder=False, verbosity=0)
fit = model.fit(X_train, y_train)

y_pred = model.predict(X_test)
predictions = [value for value in y_pred]

accuracy = accuracy_score(y_test, predictions)
print('Score:', accuracy)

plot_confusion_matrix(
    model,
    X_test,
    y_test,
    display_labels=y.map({0: 'Loss', 1: 'Win'}).unique(),
    cmap='summer_r'
)
plt.title('XGBoost Confusion Matrix')
plt.grid(False)
plt.xticks(rotation=0)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

print(knn.score(X_test, y_test))
plot_confusion_matrix(
    knn,
    X_test,
    y_test,
    display_labels=y.map({0: 'Loss', 1: 'Win'}).unique(),
    cmap='summer_r'
)
plt.title('XGBoost Confusion Matrix')
plt.grid(False)
plt.xticks(rotation=0)
plt.show()
```

```python
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB(binarize=0.0)
bnb.fit(X_train, y_train)

print(bnb.score(X_test, y_test))
plot_confusion_matrix(
    bnb,
    X_test,
    y_test,
    display_labels=y.map({0: 'Loss', 1: 'Win'}).unique(),
    cmap='summer_r'
)
plt.title('XGBoost Confusion Matrix')
plt.grid(False)
plt.xticks(rotation=0)
plt.show()
```

```python
from pprint import pprint
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [100, 1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['linear']} 
clf = GridSearchCV(SVC(random_state=42), param_grid, verbose=1, cv=5)
clf.fit(X_train, y_train)
pprint(clf.best_params_)
pprint(clf.score(X_test, y_test))
display(pd.DataFrame(list(zip(X.columns, abs(clf.best_estimator_.coef_[0]))), columns=['feature', 'percent_importance']).sort_values('percent_importance', ascending=False).head(10))
plot_confusion_matrix(
    clf,
    X_test,
    y_test,
    display_labels=y.map({0: 'Loss', 1: 'Win'}).unique(),
    cmap='summer_r'
)
plt.title('XGBoost Confusion Matrix')
plt.grid(False)
plt.xticks(rotation=0)
plt.show()
```
